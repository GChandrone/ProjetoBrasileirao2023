# Imports

```python
import pandas as pd
import requests
```

# Requisição das tabela

```python
# Pegando a url da wikipedia, onde tem várias tabelas com diversas informações sobre o brasileirão 2023
url = "https://pt.wikipedia.org/wiki/Campeonato_Brasileiro_de_Futebol_de_2023_-_S%C3%A9rie_A"

# Retorna o html do site completo
requisicao = requests.get(url)

# Retorna todas as tabelas desse html
tabelas = pd.read_html(requisicao.text)

# Escolhendo só a tabela de classificação e jogos
tabela_classificacao = tabelas[6]
tabela_jogos = tabelas[7]

display(tabela_classificacao)
display(tabela_jogos)
```

# Tratamento da Tabela de Jogos

```python
# Criando uma lista com os nomes dos times da [tabela_jogos] que estão na coluna ["Casa \ Fora"]
nomes_times = list(tabela_jogos["Casa \ Fora"])

print(nomes_times)
```

```python
# Criando uma lista com os apelidos dos times da [tabela_jogos] que estão nas [colunas]
apelidos_times = list(tabela_jogos.columns)

# Excluindo o primeiro item da lista "Casa \ Fora"
apelidos_times.pop(0)

print(apelidos_times) 
```

```python
# Criando um dicionário "Apelido: nome do time"
de_para_times = dict(zip(apelidos_times, nomes_times))

print(de_para_times)
```

```python
# Colocando o indice da tabela a coluna "Casa \ Fora"
tabela_jogos_ajustada = tabela_jogos.set_index("Casa \ Fora")

display(tabela_jogos_ajustada)
```

```python
# Transformando as colunas da tabela em linhas e ajustando o indice
tabela_jogos_ajustada = tabela_jogos_ajustada.unstack().reset_index()
display(tabela_jogos_ajustada)
```

```python
# Mudando nomes das colunas
tabela_jogos_ajustada = tabela_jogos_ajustada.rename(columns={"level_0": "Fora"})\
                                             .rename(columns={"Casa \ Fora": "Casa"})\
                                             .rename(columns={0: "Resultado"})
display(tabela_jogos_ajustada)
```

```python
# Função para trocar o apelido dos times pelo nome
def ajustar_apelido_time(linha):
    apelido = linha["Fora"]
    nome = de_para_times[apelido]
    return nome

# Aplicando a função em cada linha da [tabela_jogos_ajustada] na coluna "Fora"
tabela_jogos_ajustada["Fora"] = tabela_jogos_ajustada.apply(ajustar_apelido_time, axis=1)

#axis=0 -> coluna
#axis=1 -> linha

display(tabela_jogos_ajustada)
```

```python
# Filtrando a [tabela_jogos_ajustada] trazendo somente as linhas onde as colunas "Fora" e "Casa" forem diferentes
tabela_jogos_ajustada = tabela_jogos_ajustada[tabela_jogos_ajustada["Fora"] != tabela_jogos_ajustada["Casa"]]

display(tabela_jogos_ajustada)
```

```python
# Trocando os valores nulos por "Não aconteceu", pois são jogos que ainda vão ser disputados
tabela_jogos_ajustada["Resultado"] = tabela_jogos_ajustada["Resultado"].fillna("Não aconteceu")

# Criando uma tabela de jogos realizados. Filtrando as linhas da coluna "Resultado" que contenham o "–"
tabela_jogos_realizados = tabela_jogos_ajustada[tabela_jogos_ajustada["Resultado"].str.contains("–")]

# Criando uma tabela de jogos realizados. Filtrando as linhas da coluna "Resultado" que NÃO contenham o "–"
tabela_jogos_faltantes = tabela_jogos_ajustada[~tabela_jogos_ajustada["Resultado"].str.contains("–")]

# Trocando os valores "a" por "Não aconteceu"
tabela_jogos_faltantes.loc[tabela_jogos_faltantes["Resultado"] == "a", "Resultado"] = "Não aconteceu"

tabela_jogos_faltantes = tabela_jogos_faltantes.drop(columns=["Resultado"])

display(tabela_jogos_realizados)

display(tabela_jogos_faltantes)
```

```python
# Criando 2 colunas "Gols_Casa" e "Gols_Fora" a partir da coluna "Resultado" separada pelo "–"
tabela_jogos_realizados[["Gols_Casa", "Gols_Fora"]] = tabela_jogos_realizados["Resultado"].str.split("–", expand=True)

# Dropando a coluna "Resultado"
tabela_jogos_realizados = tabela_jogos_realizados.drop(columns=["Resultado"])

display(tabela_jogos_realizados)
```

```python
# Transformando as colunas "Gols_Fora" e "Gols_Casa" em int
tabela_jogos_realizados["Gols_Fora"] = tabela_jogos_realizados["Gols_Fora"].astype(int)
tabela_jogos_realizados["Gols_Casa"] = tabela_jogos_realizados["Gols_Casa"].astype(int)

# Criando a tabela com a média de gols marcados e gols sofridos dentro de casa
media_gols_casa = tabela_jogos_realizados.groupby("Casa").mean(numeric_only=True)
media_gols_casa = media_gols_casa.rename(columns={"Gols_Casa": "Gols_Marcados_Casa", "Gols_Fora": "Gols_Sofridos_Casa"})

# Criando a tabela com a média de gols marcados e gols sofridos fora de casa
media_gols_fora = tabela_jogos_realizados.groupby("Fora").mean(numeric_only=True)
media_gols_fora = media_gols_fora.rename(columns={"Gols_Casa": "Gols_Sofridos_Fora", "Gols_Fora": "Gols_Marcados_Fora"})

display(media_gols_casa)
display(media_gols_fora)
```

```python
# Juntando as tabelas [media_gols_casa] e [media_gols_fora], passando como paramêtro que o indice das tabelas são iguais
tabela_estatisticas = media_gols_casa.merge(media_gols_fora, left_index=True, right_index=True)

# Mudando ordem das colunas
tabela_estatisticas = tabela_estatisticas[["Gols_Marcados_Casa", "Gols_Sofridos_Casa", "Gols_Marcados_Fora", "Gols_Sofridos_Fora"]]

display(tabela_estatisticas)
```

```python
# Transformando o indice (Casa) em uma coluna da tabela
tabela_estatisticas = tabela_estatisticas.reset_index()

# Renomeando a coluna Casa para Times
tabela_estatisticas = tabela_estatisticas.rename(columns={"Casa":"Times"})

display(tabela_estatisticas)
```

# Probabilidade

```python
# Como a a partir das medias calcular a probabilidade do time ganhar/perder/empatar
time_casa = "Cruzeiro"
time_fora = "Vasco"


# [Média de gols que o Cruzeiro faz dentro de casa] * [Média de gols que o Vasco sofre dentro de casa]

#                           loc = localizar[          linha           ]==condição, "Coluna"                       
lambda_gols_casa = (tabela_estatisticas.loc[tabela_estatisticas["Times"]==time_casa, "Gols_Marcados_Casa"].iloc[0]    # Vai na coluna de "Times" onde a linha for igual a "Cruzeiro" e pega o valor da coluna "Gols_Marcados_Casa"
                                                                                                                     #                                                   * -> multiplicando
                   * tabela_estatisticas.loc[tabela_estatisticas["Times"]==time_fora, "Gols_Sofridos_Fora"].iloc[0])  # Vai na coluna de "Times" onde a linha for igual a "Vasco" e pega o valor da coluna "Gols_Sofridos_Fora"

# [Média de gols que o Cruzeiro faz dentro de casa] * [Média de gols que o Vasco sofre dentro de casa]
lambda_gols_fora = (tabela_estatisticas.loc[tabela_estatisticas["Times"]==time_fora, "Gols_Marcados_Fora"].iloc[0]   # Vai na coluna de "Times" onde a linha for igual a "Vasco" e pega o valor da coluna "Gols_Marcados_Fora"
                                                                                                                     #                                                   * -> multiplicando
                   * tabela_estatisticas.loc[tabela_estatisticas["Times"]==time_casa, "Gols_Sofridos_Casa"].iloc[0]) # Vai na coluna de "Times" onde a linha for igual a "Cruzeiro" e pega o valor da coluna "Gols_Sofridos_Casa"
```
